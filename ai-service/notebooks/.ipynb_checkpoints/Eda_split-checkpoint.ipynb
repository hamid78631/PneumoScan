## Lister les fichiers et extraires les labels + patientID

```python
import re
from pathlib import Path

def list_images(base_dir):
    paths, groups, labels = [], [], []

    for split in ['train', 'val', 'test']:
        for label in ['NORMAL', 'PNEUMONIA']:
            folder = Path(base_dir) / split / label
            for img_path in folder.glob('*.jpeg'):
                match = re.match(r"(person\d+|NORMAL2-IM-\d+|IM-\d+)", img_path.name)
                if match is None:
                    raise ValueError(f"Nom de fichier non reconnu : {img_path.name}")
                patient_id = match.group(1)
                paths.append(img_path)
                labels.append(label)
                groups.append(patient_id)
    return paths, labels, groups
```

```python
import numpy as np
from sklearn.model_selection import GroupShuffleSplit

paths, labels, groups = list_images(r"C:\Users\DELL\PneumoScan\data\chest_xray")
paths = np.array(paths)
labels = np.array(labels)
groups = np.array(groups)

# 1er split : on isole test (15%) du reste (85%)
splitter_test = GroupShuffleSplit(n_splits=1, test_size=0.15, random_state=42)
remainder_idx, test_idx = next(splitter_test.split(paths, labels, groups=groups))

remainder_paths, remainder_labels, remainder_groups = paths[remainder_idx], labels[remainder_idx], groups[remainder_idx]
test_paths, test_labels, test_groups = paths[test_idx], labels[test_idx], groups[test_idx]

# 2e split : sur le reste, on isole val (15% du reste) et train (85% du reste)
splitter_val = GroupShuffleSplit(n_splits=1, test_size=0.15, random_state=42)
train_idx, val_idx = next(splitter_val.split(remainder_paths, remainder_labels, groups=remainder_groups))

train_paths, train_labels, train_groups = remainder_paths[train_idx], remainder_labels[train_idx], remainder_groups[train_idx]
val_paths, val_labels, val_groups = remainder_paths[val_idx], remainder_labels[val_idx], remainder_groups[val_idx]

print("split effectué")
```

```python
from collections import Counter

print("Tailles -> train:", len(train_paths), "val:", len(val_paths), "test:", len(test_paths))
print("Labels train:", Counter(train_labels))
print("Labels val:", Counter(val_labels))
print("Labels test:", Counter(test_labels))

train_set, val_set, test_set = set(train_groups), set(val_groups), set(test_groups)
print("Chevauchement patients train/val:", len(train_set & val_set))
print("Chevauchement patients train/test:", len(train_set & test_set))
print("Chevauchement patients val/test:", len(val_set & test_set))
```

# Dimensions des images 


```python
from PIL import Image 
from collections import Counter 

sizes=[]
modes=[]

for p in train_paths: 
    img = Image.open(p)
    sizes.append(img.size)
    modes.append(img.mode)

print("Modes rencontrées : ", Counter(modes))
print("Nombre de tailles rencontrées : " , len(set(sizes)))
print("5 tailles les plus fréquentes : " , Counter(sizes).most_common(5))
```

## Redimensionnement des images en 64x64 , uniformisation des images au niveau gris , apatissement des images

```python
%%time
import numpy as np 
def preprocess_images(paths , size=(64, 64)): 
    vectors = []
    for p in paths : 
        img = Image.open(p).convert("L") #Conversion en gris 
        img = img.resize(size) #Redimensionnement 
        arr = np.array(img , dtype=np.float32)
        arr = arr/255 #Normaliser les pixes entre 0 et 1
        vectors.append(arr.flatten()) 
    return np.array(vectors)

X_train = preprocess_images(train_paths)
X_val = preprocess_images(val_paths) 
X_test = preprocess_images(test_paths)

y_train = (train_labels == "PNEUMONIA").astype(int)
y_val = (val_labels == "PNEUMONIA").astype(int)
y_test = (test_labels == "PNEUMONIA").astype(int)

print(X_train.shape, y_train.shape)
print("traitement terminé")
```

## Entrainement du modèle avec logisticLinearRegression 


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score , recall_score , confusion_matrix , roc_auc_score 

model = LogisticRegression(max_iter = 1000 , random_state = 42 )
model.fit(X_train , y_train )

y_pred = model.predict(X_val)
y_proba = model.predict_proba(X_val)[:,1]



print("Accuracy :", accuracy_score(y_val, y_pred))
print("Precision :", precision_score(y_val, y_pred))
print("Recall :", recall_score(y_val, y_pred))
print("ROC-AUC :", roc_auc_score(y_val, y_proba))
print("Matrice de confusion :\n", confusion_matrix(y_val, y_pred))
```
